```python
def add_scalebar(ax, x0, y0, step = 50*1000, num = 4, colors=['k','w'], size=12, lw = 3, 
                 text_offset_top = 3000, text_offset_bottom = 8000):

    x1 = x0 - step - step/1000
    if num%2==0:
        x2 = x0 + ((num-1)*step) + step/1000  
    else:
        x2 = x0 + ((num-1)*step) 
    ax.plot([x1, x2], [y0, y0], lw=lw+1, c=colors[0], zorder=100)
    
    for ii in range(num+1):
        
        
        x1 = x0 + (step*(ii-1))
        x2 = x0 + step*ii
        
        if ii < num:
            ax.plot([x1, x2], [y0, y0], lw=lw, c=colors[ii%2], zorder=100)
        
        
    
        dist = f'{step*ii/1000:.0f}'
        ax.text(x1, y0+text_offset_top, dist, va='bottom', ha = 'center', size=size, zorder=100)
    
    
    ax.text(x0 + ((num/2-1)*step), y0-text_offset_bottom, 'Kilometers', va='top', ha = 'center', size=size, zorder=100)
```

```python


#0=clouds, 1=land, 2=sea ice, 3=artefacts, 4=leads, 5=open water.
lead_cmap = matplotlib.colors.ListedColormap(['lightgray', 'gray', 'white', 'orange', 'dodgerblue', 'blue'])

def open_local_lead_file(year, crop_i = [1400, 2200], crop_j = [600, 1200]):
    
    # crop_i is range to crop i index
    # crop_j is range to crop j index
    
    # JFMA year corresponds to second year of file
    # Nov. year1 -- Apr. year2
    
    # open data file
    #0=clouds, 1=land, 2=sea ice, 3=artefacts, 4=leads, 5=open water.
    da = xr.open_dataset(f'/Volumes/Seagate_Jewell/KenzieStuff/ArcLeads/{year-1}{str(year)[-2:]}_ArcLeads.nc')
    da.close()
    
    # interpret file dates
    # note that seocond set of dates (those after Jan 1) list the wrong (previous) year
    dates1 = [datetime.strptime(time, '%Y%m%d') for time in da.time.values[:61].astype(str)]
    dates2 = [datetime(int(time[:4])+1, int(time[4:6]), int(time[6:])) for time in da.time.values[61:].astype(str)]
    dates = np.append(dates1, dates2)
    
    # crop
    ai, bi = crop_i[0], crop_i[1]
    aj, bj = crop_j[0], crop_j[1]
    
    # grab coordinates
    lats = da.lat[ai:bi, aj:bj]
    lons = da.lon[ai:bi, aj:bj]
    
    # lea map
    lead_map = da.leadmap[:, ai:bi, aj:bj]
    
    return lons, lats, dates, lead_map

```

```python
mean_sic = {}
open_freq = {}


for group in all_group_dates.keys():

    print(group)
    
    open_counts = np.zeros(hrsic['sic'][Li:Lj, La:Lb].shape)
    group_counts = np.zeros(hrsic['sic'][Li:Lj, La:Lb].shape)
    group_dates = pd.to_datetime(all_group_dates[group][all_group_dates[group].month<4])
    sic1 = np.zeros(hrsic['sic'][Li:Lj, La:Lb].shape)

    for date in group_dates:
        
        try:
            hrsic = UNISIC.open_local_file(date, res = '6250', hem = 'n',
                                                  area = False, include_units=False)
            local_sic = np.copy(hrsic['sic'][Li:Lj, La:Lb])
            
            if np.sum(np.isfinite(local_sic)) > 1000:
                
                group_counts[np.isfinite(local_sic)] +=1
                
                open_counts[local_sic<70] += 1
                
                local_sic[np.isnan(local_sic)] = 0
                sic1 += local_sic

                
            
        except:
            group_counts+=0


    mean_sic[group] = sic1 / group_counts
    open_freq[group] = open_counts / group_counts
    
    
mean_sic['xx'] = hrsic['xx'][Li:Lj, La:Lb]
mean_sic['yy'] = hrsic['yy'][Li:Lj, La:Lb]
mean_sic['proj'] = hrsic['proj']
```

```python
years = np.append(np.arange(2003,2012), np.arange(2013,2023))


large_events = {}
large_events['flaw'] = np.array([], dtype='datetime64[ns]')
large_events['lfbo'] = np.array([], dtype='datetime64[ns]')

for year in years:

    times = pd.date_range(datetime(year-1, 12, 1), datetime(year,3,31))
#     times = pd.date_range(datetime(year, 1, 1), datetime(year,3,31))
    
    open_areas = {}
    open_areas['sic'] = {}
    open_areas['sic']['inner'] = np.array([])
    open_areas['sic']['middle'] = np.array([])


    open_areas['lead'] = {}
    open_areas['lead']['inner'] = np.array([])
    open_areas['lead']['middle'] = np.array([])

    for date in times:

        # open sic data
        try:
            data['sic'] = UNISIC.open_local_file(date, res='6250', hem='n', 
                                                 main_path='/Volumes/Seagate_Jewell/KenzieStuff/', 
                                                 coordinates=False, area=False, include_units=False, quiet=True)
        except:
            data['sic'] = {}
            data['sic']['sic'] = np.full(lfbo_range.shape, np.nan)
        
        inner_area = np.sum(data['sic']['sic'][lfbo_range] < 70) * 6.25*6.25
        middle_area = np.sum(data['sic']['sic'][flaw_range] < 70) * 6.25*6.25

        open_areas['sic']['inner'] = np.append(open_areas['sic']['inner'], inner_area)
        open_areas['sic']['middle'] = np.append(open_areas['sic']['middle'], middle_area)


        # open lead file
        #0=clouds, 1=land, 2=sea ice, 3=artefacts, 4=leads, 5=open water.
        if date.month == 12:
            YEAR = date.year+1
        else:
            YEAR = date.year
            
        lons, lats, dates, lead_map = open_local_lead_file(YEAR, 
                                                           crop_i = [1590, 1855], 
                                                           crop_j = [780, 980])


        new_leads = np.full(lats.shape, False)

        # open specific dates:
        #----------------------
        # current day's map
        tind = np.where(dates == date)[0]
        if len(tind) == 0:
            lons, lats, dates, lead_map = open_local_lead_file(date.year+1, 
                                                               crop_i = [1590, 1855], 
                                                               crop_j = [780, 980])
            tind = np.where(dates == date)[0]
            
        curr_map = lead_map[tind[0]]
        curr_map = check_unusual(curr_map)               
        curr_leads = (curr_map >= 4).astype(int)
        curr_cloud = (curr_map == 0).astype(int)

        # previous day's map
        tind = np.where(dates == date - timedelta(days=1))[0]
        prev_map = lead_map[tind[0]]
        prev_map = check_unusual(prev_map)    
            
        prev_ice = (prev_map == 2).astype(int)
        prev_leads = (prev_map >= 4).astype(int)
        prev_cloud = (prev_map == 0).astype(int)

        # 2 days previous map
        tind = np.where(dates == date + timedelta(days=1))[0]
        next_map = lead_map[tind[0]]
        next_map = check_unusual(next_map)   
        
        next_leads = (next_map >= 4).astype(int)

        # following day's map
        tind = np.where(dates == date - timedelta(days=2))[0]
        prev2_map = lead_map[tind[0]]
        prev2_map = check_unusual(prev2_map)   
        
        prev2_ice = (prev2_map == 2).astype(int)


        # include different combos as active leads
        #-----------------------------------------

        # current leads that were ice the previous day
        new_leads[(curr_leads+prev_ice==2)] = True

        # current leads that were cloud-covered previous day, but ice the day before that
#         new_leads[(curr_leads+prev_cloud+prev2_ice==3)] = True

        # current clouds that were ice-covered previous day, but lead the following day
        new_leads[(curr_cloud+prev_ice+next_leads==3)] = True

        # include leads alive for one day
        # current leads that were also leads previous day, but ice the day before that
        new_leads[(curr_leads+prev_leads+prev2_ice==3)] = True



        inner_area = np.sum(new_leads[inner])
        middle_area = np.sum(new_leads[middle])

        open_areas['lead']['inner'] = np.append(open_areas['lead']['inner'], inner_area)
        open_areas['lead']['middle'] = np.append(open_areas['lead']['middle'], middle_area)


    avg_flaw = (0.5*open_areas['lead']['middle'] + 0.5*open_areas['sic']['middle'])
    avg_lfbo = (0.5*open_areas['lead']['inner']  + 0.5*open_areas['sic']['inner'])
        
    fig, ax = plt.subplots(figsize=(8,2))
    plt.plot(times, open_areas['lead']['middle'], lw=1, ls='dashed', c='orange')
    plt.plot(times, open_areas['lead']['inner'], lw=1, ls='dashed', c='b')
    plt.plot(times, open_areas['sic']['middle'], lw=1, c='orange')
    plt.plot(times, open_areas['sic']['inner'], lw=1, c='b')
    plt.plot(times, avg_flaw, lw=3, c='orange', alpha=0.5, zorder=0)
    plt.plot(times, avg_lfbo, lw=3, c='b', alpha=0.5, zorder=0)
    
    ax.axhline(500, c='k', zorder=-1)
    ax.set_ylim(-100,3000)
    plt.show()

    # save dates
    large_flaw = (avg_flaw > 500) & (open_areas['sic']['middle'] > 0)
    increasing = np.append(np.array([False]), np.diff(avg_flaw)>0)
    large_events['flaw'] = np.append(large_events['flaw'], times[increasing&large_flaw])

    large_lfbo = (avg_lfbo > 500) & (open_areas['sic']['inner'] > 0)
    increasing = np.append(np.array([False]), np.diff(avg_lfbo)>0)
    large_events['lfbo'] = np.append(large_events['lfbo'], times[increasing&large_lfbo])
    
large_events['lfbo'] = pd.to_datetime(large_events['lfbo'])
large_events['flaw'] = pd.to_datetime(large_events['flaw'])

large_events['lfbo_only'] = np.array([date for date in large_events['lfbo'] if date not in large_events['flaw']])
large_events['flaw_only'] = np.array([date for date in large_events['flaw'] if date not in large_events['lfbo']])
large_events['both'] = np.array([date for date in large_events['flaw'] if date in large_events['lfbo']])


```

```python
def calc_open_freq_sic(dates, ai,aj, bi,bj):

    counter = 0
    
    for date in dates:

        # open local sic file
        try:
            sic = UNISIC.open_local_file(date, res = '6250', hem = 'n',
                                          area = False, include_units=False)
            exists = True
            counter+=1
        except:
            exists = False

        if counter == 1:
            
            # mean
            #-------------------------------------------------
            sic_sum = np.zeros(sic['sic'][ai:aj, bi:bj].shape)
            open_sum = np.zeros(sic['sic'][ai:aj, bi:bj].shape)
            non_nan = np.zeros(sic['sic'][ai:aj, bi:bj].shape)
            
        if exists:

            sic_data = np.copy(sic['sic'][ai:aj, bi:bj])

            # record non-nan values, then replace nans with zeros
            non_nan += np.isfinite(sic_data).astype(int)
            sic_data[np.isnan(sic_data)] = 0
            sic_sum += sic_data
            open_sum += (sic_data < 70).astype(int)

        # record progress
        if counter%100 == 0:
            print(f'[{counter}/{len(dates)}] {date}')

    sic_mean = sic_sum/non_nan
    open_freq = open_sum/non_nan
    print('---- mean calculated')

    return sic_mean, open_freq



# # standard deviation
# #-------------------------------------------------
# sic_sum = np.zeros(data['sic']['sic'][ai:aj, bi:bj].shape)
# non_nan = np.zeros(data['sic']['sic'][ai:aj, bi:bj].shape)

# for date in dates:

#     # open local sic file
#     try:
#         sic = UNISIC.open_local_file(date, res = '6250', hem = 'n',
#                                       area = False, include_units=False)
#         exists = True
#     except:
#         exists = False
        
        
#     if exists:
        
#         sic_data = (sic_mean - np.copy(sic['sic'][ai:aj, bi:bj]))**2

#         # record non-nan values, then replace nans with zeros
#         non_nan += np.isfinite(sic_data).astype(int)
#         sic_data[np.isnan(sic_data)] = 0
#         sic_sum += sic_data

# sic_std = np.sqrt(sic_sum/non_nan)
```
